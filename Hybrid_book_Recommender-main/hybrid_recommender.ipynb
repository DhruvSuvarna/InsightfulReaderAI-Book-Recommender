## Collaborative Filtering

```python
x= new_books.groupby('User-ID').count()['Book-Rating'] >= 1
padhe_likhe_users=x[x].index
```

```python
filtered_rating = new_books[new_books['User-ID'].isin(padhe_likhe_users)]
filtered_rating
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating'] >=1
famous_books=y[y].index
famous_books.duplicated().sum()
famous_books
```

```python
final_ratings= filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings.drop_duplicates()
```

```python
pt= final_ratings.pivot_table(index='Book-Title', columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt.index
```

```python
book_titles = pt.index

# Specify the file path
file_path = "book_titles.txt"

# Open the file in write mode and save the titles
with open(file_path, "w") as file:
    for title in book_titles:
        file.write(title + "\n")

print("Book titles saved to:", file_path)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores=cosine_similarity(pt)
```

```python
similarity_scores
```

```python
def collab_filt_recommend(book_name):
    #index fetch
    index = np.where(pt.index == book_name)[0][0]
    
    similar_items= sorted(list(enumerate(similarity_scores[index])),key = lambda x:x[1],reverse=True)[1:6]
    L=[]
    for i in similar_items:
        L.append(pt.index[i[0]])
    return L
```

```python
collab_list=collab_filt_recommend("Charlotte's Web")
collab_list
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words="english")
cv
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
len(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[11])),reverse = True, key=lambda x :x[1])[1:6]
```

```python
def content_based_recommend(book):
    
    book_index = new_df[new_df['Book'] == book].index[0]
    distances = similarity[book_index]
    book_list= sorted(list(enumerate(distances)),reverse = True, key=lambda x :x[1])[1:6]
    L=set()
    for i in book_list:
       
        L.add(new_df.iloc[i[0]].Book)
    return L
```

```python
c_books=content_based_recommend("Charlotte's Web")
c_books
```

```python
def hybrid_recommender(book_recommend):

    final_list = []
    try:
      content_books = content_based_recommend(book_recommend) 
      print("Content",content_books)
      for i in content_books:
        item={'title':'', 'author':'', 'url':'', 'rating': '', 'genres':[],'description': ''}
        temp_df=book[book['Book'] == i]
           
        temp_df_isbn=temp_df['ISBN']
        for isbn in temp_df_isbn:
         isbn = isbn.strip()  # Remove leading and trailing whitespace
         url = f'https://covers.openlibrary.org/b/isbn/{isbn}-M.jpg'
         item['url']=url
         
        # print(item['url'])
        item['title'] = str(temp_df.drop_duplicates('Book')['Book'].values).replace("['", "").replace("']", "")
        item['author'] = str(temp_df.drop_duplicates('Book')['Author'].values).replace("[list(['", "").replace("'])]", "")
        item['rating'] = str(temp_df.drop_duplicates('Book')['Avg_Rating'].values).replace("[", "").replace("]", "")
        # item['url'] = str(temp_df.drop_duplicates('Book')['Image-URL-L'].values).replace("['", "").replace("']", "")
        item['genres'] = list(str(temp_df.drop_duplicates('Book')['Genres'].values).replace('["', '').replace('"]', '').replace("[", "").replace("]", "").replace('\\', '').replace("'", "").replace("list(", "").replace(")", "").split(', '))
        item['description'] = str(temp_df.drop_duplicates('Book')['Description'].values).replace('["', '').replace('"]', '').replace("[", "").replace("]", "").replace('\\', '').replace("'", "").replace("list(", "").replace(")", "").replace(",","")
        print(item)
        final_list.append(item)
    except Exception as e:
      print("Something went wrong when writing to the file",e)
      content_books=[]
    
    try:
        collab_books = collab_filt_recommend(book_recommend)
        print("collab",collab_books)
        for i in collab_books:
          item={'title':'', 'author':'', 'url':'','genres':'','description': ''}
          temp_df=new_books[new_books['Book-Title'] == i]
          temp_df_isbn=temp_df['ISBN']
          for isbn in temp_df_isbn:
           isbn = isbn.strip()  # Remove leading and trailing whitespace
           url = f'https://covers.openlibrary.org/b/isbn/{isbn}-M.jpg'
           item['url']=url
         
          item['title'] = str(temp_df.drop_duplicates('Book-Title')['Book-Title'].values).replace("['", "").replace("']", "")
          item['author'] = str(temp_df.drop_duplicates('Book-Title')['Book-Author'].values).replace("[list(['", "").replace("'])]", "")
          item['rating'] = str(temp_df.drop_duplicates('Book')['Avg_Rating'].values).replace("[", "").replace("]", "")
          # item['url'] = str(temp_df.drop_duplicates('Book-Title')['Image-URL-L'].values).replace("['", "").replace("']", "")
          item['genres'] = list(str(temp_df.drop_duplicates('Book')['Genres'].values).replace('["', '').replace('"]', '').replace("[", "").replace("]", "").replace('\\', '').replace("'", "").replace("list(", "").replace(")", "").split(', '))
          item['description'] = str(temp_df.drop_duplicates('Book')['Description'].values).replace('["', '').replace('"]', '').replace("[", "").replace("]", "").replace('\\', '').replace("'", "").replace("list(", "").replace(")", "").replace(",","")
          print(item)
          final_list.append(item) 
    except:
       print("Something wrong in collab")
       collab_books=[]
    
    return final_list
```

```python
from flask import Flask, request, jsonify

app = Flask(__name__)

book_list = []


@app.route('/books', methods=['GET', 'POST'])
def manage_books():

    if request.method == 'GET':
        if book_list:
            return jsonify(book_list)  # Return all books if list is not empty
        else:
            return jsonify({'message': 'No books found'}), 404  # Clearer message

    elif request.method == 'POST':

        title = request.form['title']
        book_list = hybrid_recommender(title)

        return jsonify(book_list)  # Created response with the new book

if __name__ == '__main__':
    app.run(host='localhost')
```
